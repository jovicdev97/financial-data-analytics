```python
#Anforderung 4
import pandas as pd

# Load Dataset
df = pd.read_csv('Loan.csv')

# Show first 5
df.head()
```

```python
# Anforderung 1
import numpy as np

# Integer array
int_array = np.array([25, 35, 45, 55, 65], dtype=int)

# Float array
float_array = np.array([0.45, 1.23, 3.67, 2.89, 4.56], dtype=float)

# String array
string_array = np.array(['employed', 'unemployed', 'self-employed', 'retired', 'student'], dtype=str)

# Boolean array
bool_array = np.array([True, False, True, False, True], dtype=bool)

# Display the manually created arrays
int_array, float_array, string_array, bool_array
```

```python
# Anforderung 2
# 1. Integer array values greater than 30
# 2. Boolean array is True

filtered_int_array = int_array[(int_array > 30) & bool_array]

# Display the filtered arrays
filtered_int_array
```

```python
#Anforderung 3

# For the integer array
mean_int_array = np.mean(int_array)
median_int_array = np.median(int_array)

# For the float array
max_float_array = np.max(float_array)
min_float_array = np.min(float_array)

# Displaying the results for the float array
mean_int_array, median_int_array, max_float_array, min_float_array
```

```python
""" Anforderung 5: Monthly Income / Monthly Loan Payment """
""" Erstelle neue Zeile im bestehenden Dataaset df "Neue_Zeile_SCHULDENVERHÃ„LTNIS """

df["Neue_Zeile_SCHULDENVERHÃ„LTNIS"] = (
    df["MonthlyIncome"] / df["MonthlyLoanPayment"]
)

""" Adde neue Spalte zum PANDA DATAFRAME DF """
df["Neue_Zeile_SCHULDENVERHÃ„LTNIS"]
```

```python
df.head()
```

```python
# Change name of "Neue_Zeile_SCHULDENVERHÃ„LTNIS" to Spalte (weil es ist eien Spalte)
# Anforderung 6
df.rename(columns={"Neue_Zeile_SCHULDENVERHÃ„LTNIS": "SchuldenverhÃ¤ltnis"}, inplace=True)

# switch index -1 column with index 0 column
cols = df.columns.tolist()
cols = [cols[-1]] + cols[:-1]
df = df[cols]

df.head()
```

```python
""" Anforderung 7 : SPalte lÃ¶schen """
df = df.drop('SchuldenverhÃ¤ltnis', axis=1) 
df.head()
```

```python
""" Anforderung 8 """
""" einmal toLower einmal toUpper """
df["EmploymentStatus"] = df["EmploymentStatus"].str.lower()
df.head()
df["EmploymentStatus"] = df["EmploymentStatus"].str.upper()
df.head()
```

```python
""" Anforderung 9 """
df.to_csv('Loan_neu.csv', index=False)
```

```python
""" Diagramm 1 - Liniendiagramm """

import matplotlib.pyplot as plt
import seaborn as sns

fig, ax = plt.subplots(figsize=(10,6))


ax.set_title("Zinssatz nach Jahr")
ax.set_xlabel("Jahr")
ax.set_ylabel("Zinsfaktor")
plt.ylim(0.2, 0.28) #Skalierung (y-achse)
df['ApplicationDate'] = pd.to_datetime(df['ApplicationDate'])
df_sorted_year = df.sort_values(by='ApplicationDate')
df['Year'] = df['ApplicationDate'].dt.year # "Herausziehen" von Jahren aus dem ApplicationDate
yearly_avg_interest = df.groupby('Year')['BaseInterestRate'].mean()#Gruppierung von BaseInterestRate nach Jahr

ax.plot(yearly_avg_interest.index, yearly_avg_interest.values)


plt.show()

#Erkenntnisse:
#Datenstruktur          : Linien Diagramm eignet sich am besten bei zeitlichen AblÃ¤ufen
#Daten                  : Daten bleiben Ã¼ber den zeitlichen Verlauf nahezu konstant
#GeschÃ¤ftliche DomÃ¤ne   : Zeitlicher Verlauf spielt keine groÃŸe Rolle fÃ¼r den Zinsfaktor
```

```python
"""Diagramm 2 - Balkendiagramm"""

employment_counts = df["EmploymentStatus"].value_counts() #ZÃ¤hle EintrÃ¤ge der verschiedenen Employment Statuse und speicher diese in einer Serie
employment_counts.plot(kind="bar") #Generiere Diagramm aus Serie
plt.xlabel("BeschÃ¤ftigungsstatus")
plt.ylabel("Anzahl der Antragsteller")
plt.title("Anzahl der Antragsteller nach BeschÃ¤ftigungsstatus")
plt.show()

#Erkenntnisse:
#Datenstruktur          : Balkendiagramm eignet sich fÃ¼r wenige und klar definierbare Merkmale
#Daten                  : HauptsÃ¤chlich Angestellte
#GeschÃ¤ftliche DomÃ¤ne   : Angestellte haben klaren Vorang bei Optimierung Kreditvergabe
```

```python
"""Diagramm 3 - Histogramm"""

df["AnnualIncome"].plot(kind="hist", bins=30) #edgecolor = 'k' sorgt fÃ¼r Umrandung
plt.xlabel("JÃ¤hrliches Einkommen in USD")
plt.ylabel("HÃ¤ufigkeit")
plt.title("Verteilung des jÃ¤hrlichen Einkommens in USD")
plt.show()

#Erkenntnisse:
#Datenstruktur          : Histogramm eignet sich um eine breite Verteilung von Werten darzustellen
#Daten                  : JÃ¤hrliches Einkommen ist hauptsÃ¤chlich unter 100.000 USD
#GeschÃ¤ftliche DomÃ¤ne   : Die meisten Antragsteller haben ein jÃ¤hrliches Einkommen von unter 100.000 USD
```

```python
"""Diagramm 4 - BoxPlot"""

# Definieren der Reihenfolge der Bildungsgrade
education_order = ["High School", "Associate", "Bachelor", "Master", "Doctorate"]
# Konvertieren der 'EducationLevel'-Spalte in eine kategorische Datenstruktur
df["EducationLevel"] = pd.Categorical(
    df["EducationLevel"], categories=education_order, ordered=True
)
# DataFrame nach 'EducationLevel' sortieren
df_sorted = df.sort_values(by="EducationLevel")
# Boxplot erstellen
boxplot = df_sorted.boxplot(column="BaseInterestRate", by="EducationLevel", grid=True)
plt.xlabel("Bildungsgrad X-Achse")
plt.ylabel("Basis Zinsrate in %")
plt.show()


#Erkenntnisse:
#Datenstruktur          : Boxplot eignet sich um AusreiÃŸer direkt zu erkennen und Quartile zu beobachten
#Daten                  : Zinsrate ist in Bezug auf den Bildungsgrad relativ gleichmÃ¤ÃŸig verteilt
#GeschÃ¤ftliche DomÃ¤ne   : BeschÃ¤ftigungsgrad hat keinen groÃŸen Einfluss auf die Zinsrate
```

```python
"""Diagramm 5 - ScatterPlot """

plt.figure(figsize=(10, 6)) 
plt.scatter(df['CreditScore'], df['BaseInterestRate'], alpha=0.5) #Alpha Ã¤ndert die Transparenz der Punkte
plt.title("Basis Zinsrate vs Credit Score")
plt.xlabel("Credit Score") 
plt.ylabel("Basis Zinsrate in % ") 
plt.grid(True) 
plt.show()

#ToDo: Daten vor Abbildung sortieren

#Erkenntnisse:
#Datenstruktur          : Scatterplot eignet sich gut um eine Verteilung und Streuung der Daten zu Visulaisieren. Zudem lassen sich AusreiÃŸer gut erkennen 
#Daten                  : Bei einem hÃ¶heren Credit Score sinkt die Zinsrate
#GeschÃ¤ftliche DomÃ¤ne   : Je hÃ¶her der Credit Score, desto geringer kann die Zinsrate angesetzt werden
```
