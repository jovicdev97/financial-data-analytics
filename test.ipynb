```python
#Anforderung 4
import pandas as pd

# Load Dataset
df = pd.read_csv('Loan.csv')

# Show first 5
df.head()
```

```python
# Anforderung 1
import numpy as np

# Integer array
int_array = np.array([25, 35, 45, 55, 65], dtype=int)

# Float array
float_array = np.array([0.45, 1.23, 3.67, 2.89, 4.56], dtype=float)

# String array
string_array = np.array(['employed', 'unemployed', 'self-employed', 'retired', 'student'], dtype=str)

# Boolean array
bool_array = np.array([True, False, True, False, True], dtype=bool)

# Display the manually created arrays
int_array, float_array, string_array, bool_array
```

```python
# Anforderung 2
# 1. Integer array values greater than 30
# 2. Boolean array is True

filtered_int_array = int_array[(int_array > 30) & bool_array]

# Display the filtered arrays
filtered_int_array
```

```python
#Anforderung 3

# For the integer array
mean_int_array = np.mean(int_array)
median_int_array = np.median(int_array)

# For the float array
max_float_array = np.max(float_array)
min_float_array = np.min(float_array)

# Displaying the results for the float array
mean_int_array, median_int_array, max_float_array, min_float_array
```

```python
""" Anforderung 5: Monthly Income / Monthly Loan Payment """
""" Erstelle neue Zeile im bestehenden Dataaset df "Neue_Zeile_SCHULDENVERHÃ„LTNIS """

df["Neue_Zeile_SCHULDENVERHÃ„LTNIS"] = (
    df["MonthlyIncome"] / df["MonthlyLoanPayment"]
)

""" Adde neue Spalte zum PANDA DATAFRAME DF """
df["Neue_Zeile_SCHULDENVERHÃ„LTNIS"]
```

```python
df.head()
```

```python
# Change name of "Neue_Zeile_SCHULDENVERHÃ„LTNIS" to Spalte (weil es ist eien Spalte)
# Anforderung 6
df.rename(columns={"Neue_Zeile_SCHULDENVERHÃ„LTNIS": "SchuldenverhÃ¤ltnis"}, inplace=True)

# switch index -1 column with index 0 column
cols = df.columns.tolist()
cols = [cols[-1]] + cols[:-1]
df = df[cols]

df.head()
```

```python
""" Anforderung 7 : SPalte lÃ¶schen """
df = df.drop('SchuldenverhÃ¤ltnis', axis=1) 
df.head()
```

```python
""" Anforderung 8 """
""" einmal toLower einmal toUpper """
df["EmploymentStatus"] = df["EmploymentStatus"].str.lower()
df.head()
df["EmploymentStatus"] = df["EmploymentStatus"].str.upper()
df.head()
```

```python
""" Anforderung 9 """
df.to_csv('Loan_neu.csv', index=False)
```

```python
""" Diagramm 1 - Liniendiagramm """

import matplotlib.pyplot as plt
import seaborn as sns

fig, ax = plt.subplots(figsize=(10,6))


ax.set_title("Zinssatz nach Jahr")
ax.set_xlabel("Jahr")
ax.set_ylabel("Zinsfaktor")
plt.ylim(0.2, 0.28) #Skalierung (y-achse)
df['ApplicationDate'] = pd.to_datetime(df['ApplicationDate'])

df['Year'] = df['ApplicationDate'].dt.year # "Herausziehen" von Jahren aus dem ApplicationDate
yearly_avg_interest = df.groupby('Year')['BaseInterestRate'].mean()#Gruppierung von BaseInterestRate nach Jahr

ax.plot(yearly_avg_interest.index, yearly_avg_interest.values)


plt.show()
```
