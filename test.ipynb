```python
import pandas as pd

# Load Dataset
df = pd.read_csv('Loan.csv')

# Show first 5
df.head()
```

```python
""" Anforderung 5: Monthly Income / Monthly Loan Payment """
""" Erstelle neue Zeile im bestehenden Dataaset df "Neue_Zeile_SCHULDENVERHÄLTNIS """

df["Neue_Zeile_SCHULDENVERHÄLTNIS"] = (
    df["MonthlyIncome"] / df["MonthlyLoanPayment"]
)

""" Adde neue Spalte zum PANDA DATAFRAME DF """
df["Neue_Zeile_SCHULDENVERHÄLTNIS"]
```

```python
df.head()
```

```python
# Change name of "Neue_Zeile_SCHULDENVERHÄLTNIS" to Spalte (weil es ist eien Spalte)
# Anforderung 6
df.rename(columns={"Neue_Zeile_SCHULDENVERHÄLTNIS": "Schuldenverhältnis"}, inplace=True)

# switch index -1 column with index 0 column
cols = df.columns.tolist()
cols = [cols[-1]] + cols[:-1]
df = df[cols]

df.head()
```

```python
""" Anforderung 7 : SPalte löschen """
df = df.drop('Schuldenverhältnis', axis=1) 
df.head()
```

```python
""" Anforderung 8 """
""" einmal toLower einmal toUpper """
df["EmploymentStatus"] = df["EmploymentStatus"].str.lower()
df.head()
df["EmploymentStatus"] = df["EmploymentStatus"].str.upper()
df.head()
```

```python
""" ANforderung 9 """
df.to_csv('Loan_neu.csv', index=False)
```
